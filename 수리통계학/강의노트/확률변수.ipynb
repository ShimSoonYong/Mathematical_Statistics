# 확률의 개념
- 확률실험
- 표본공간, 사건, 확률
- 확률의 정의(공리)
- 확률의 기본 성질

확률실험은 결과를 미리 알 수 없는 실험이다.  
- 결과 자체를 모르는 게 아니라 특정 결과를 확정지을 수 없는 실험
- 예: 동전 던지기. 앞면과 뒷면 중 하나인 건 알지만 둘 중 하나임을 확정할 수는 없다.

표본공간: 확률실험의 결과 집합  
사건: 표본공간의 부분집합

__확률의 정의__  
- 0과 1 사이의 실수
- 표본공간의 확률은 1(100%)
- 여사건의 확률은 1에서 사건의 확률을 뺀 것
- 큰 집합의 확률은 작은 집합의 확률보다 크다.

극한에 대한 것은 생략. 강의자료에 없고 교수가 설명하지 않은 것도 생략.

이미지에 표시된 내용에 따라, 확률에 대한 기본 개념을 설명하겠습니다.

### 1. 확률 실험

 확률 실험(Probabilistic Experiment)  은 결과가 무작위로 나타나며, 각 결과가 나타날 가능성에 대해 특정 확률이 정의되는 실험을 말합니다. 이 실험의 결과는 예측할 수 없지만, 각 결과가 발생할 확률은 계산할 수 있습니다. 예를 들어, 주사위를 던지는 것은 대표적인 확률 실험입니다. 주사위를 던지기 전에는 어떤 숫자가 나올지 알 수 없지만, 각 숫자가 나올 확률(1/6)은 계산 가능합니다.

#### 예시:
- 동전을 던지는 실험에서, 동전이 앞면이 나올지 뒷면이 나올지 예측할 수 없지만, 각각의 확률은 1/2로 계산됩니다.

### 2. 표본 공간, 사건, 확률

#### 표본 공간 (Sample Space, $ S $):
표본 공간은  확률 실험의 모든 가능한 결과들의 집합  을 의미합니다. 어떤 실험에서 나올 수 있는 모든 결과를 모은 집합입니다.

#### 사건 (Event):
사건은 표본 공간의  부분집합  입니다. 즉, 하나 이상의 결과를 포함하는 집합으로, 확률 실험에서 발생할 수 있는 특정 결과 또는 결과들의 모음을 말합니다. 사건은 일반적으로 $ A, B, C $ 등으로 표시합니다.

#### 확률 (Probability):
확률은 특정 사건이 일어날 가능성을 숫자로 나타낸 값입니다. 확률은 항상 0과 1 사이의 값으로 주어지며, 0은 사건이 절대로 일어나지 않음을 의미하고, 1은 사건이 반드시 일어남을 의미합니다. 

$$
P(A) = \frac{\text{A 사건의 발생 가능한 경우의 수}}{\text{전체 가능한 경우의 수}}
$$

#### 예시:
- 동전을 던질 때의 표본 공간 $ S = \{앞, 뒤\} $.
- 사건 $A $: "앞면이 나온다".
- $ P(A) = \frac{1}{2} $.

### 3. 확률의 정의 (공리)

 공리적 확률 이론(Axiomatic Probability Theory)  은 수학자  콜모고로프(Andrey Kolmogorov)  가 제안한 확률의 공리적 정의로, 세 가지 기본 공리로 확률을 정의합니다.

1.  비음성 공리(Non-negativity)  : 사건 $ A $의 확률은 0 이상이어야 한다.
   $$
   P(A) \geq 0
   $$

2.  전체 확률(Entirety)  : 표본 공간에서 일어날 수 있는 모든 사건의 확률은 1이다. 즉, 반드시 어떤 사건이 발생한다.
   $$
   P(S) = 1
   $$

3.  가산적 공리(Additivity)  : 서로 배반적인(즉, 동시에 일어날 수 없는) 사건 $ A $와 $ B $의 확률은 각각의 사건 확률을 더한 값과 같다.
   $$
   P(A \cup B) = P(A) + P(B) \quad \text{(단, $ A \cap B = \emptyset $)}
   $$

이 세 가지 공리를 바탕으로, 확률이 어떻게 정의되고 계산되는지 설명할 수 있습니다.

### 4. 확률의 기본 성질

확률 이론에서 중요한 몇 가지 기본 성질들이 있습니다.

1.  확률의 범위  : 모든 사건의 확률은 0과 1 사이의 값을 가집니다.
   $$
   0 \leq P(A) \leq 1
   $$

2.  여집합의 확률  : 사건 $ A $가 일어나지 않을 확률, 즉 사건 $ A $의 여집합 $ A^c $의 확률은 $ 1 - P(A) $로 계산됩니다.
   $$
   P(A^c) = 1 - P(A)
   $$

3.  전체 사건의 확률  : 표본 공간에서 모든 결과가 발생할 확률은 항상 1입니다.
   $$
   P(S) = 1
   $$

4.  조건부 확률(Conditional Probability)  : 사건 $ B $가 일어났다는 조건 하에 사건 $ A $가 일어날 확률은 다음과 같이 정의됩니다:
   $$
   P(A \mid B) = \frac{P(A \cap B)}{P(B)} \quad \text{(단, $ P(B) \neq 0 $)}
   $$

5.  독립성(Independence)  : 두 사건 $ A $와 $ B $가 독립이라는 것은 한 사건이 일어나도 다른 사건이 일어날 확률에 영향을 미치지 않는다는 의미입니다.
   $$
   P(A \cap B) = P(A) \times P(B)
   $$

이 기본 성질들은 다양한 확률 문제를 풀 때 중요한 역할을 합니다.

### 1. 조건부 확률 (Conditional Probability)

조건부 확률은 특정 사건이 발생했음을 알고 있을 때, 다른 사건이 발생할 확률을 의미합니다. 즉, 사건 $ B $가 발생한 상황에서 사건 $ A $가 발생할 확률을 말합니다. 이는 사건 $ A $와 $ B $가 독립적이지 않을 때 중요한 개념입니다.

#### 조건부 확률의 정의:

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)} \quad \text{(단, $ P(B) \neq 0 $)}
$$

- $ P(A \mid B) $: 사건 $ B $가 발생한 후 사건 $ A $가 발생할 확률.
- $ P(A \cap B) $: 사건 $ A $와 $ B $가 동시에 발생할 확률.
- $ P(B) $: 사건 $ B $가 발생할 확률.

#### 예시:

동전을 두 번 던지는 실험을 생각해보겠습니다. 표본 공간 $ S = \{HH, HT, TH, TT\} $에서, 두 번째 동전이 앞면이 나올 확률을 구하려고 합니다. 첫 번째 동전이 앞면이 나왔다는 조건 하에서, 두 번째 동전이 앞면이 나올 확률은:

- 사건 $ A $: 두 번째 동전이 앞면이 나옴 $ \{HH, TH\} $.
- 사건 $ B $: 첫 번째 동전이 앞면이 나옴 $ \{HH, HT\} $.

이때 조건부 확률은 다음과 같습니다:

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)} = \frac{P(\{HH\})}{P(\{HH, HT\})} = \frac{\frac{1}{4}}{\frac{1}{2}} = \frac{1}{2}
$$

즉, 첫 번째 동전이 앞면이 나왔을 때, 두 번째 동전이 앞면이 나올 확률은 1/2입니다.

### 2. 베이즈 정리 (Bayes' Theorem)

베이즈 정리는 조건부 확률을 활용하여, 사건 $ B $가 발생한 후 사건 $ A $의 확률을 구하는 방법을 제공합니다. 이 정리는 기존의 조건부 확률을 뒤집어, 새로운 정보가 주어졌을 때 확률을 업데이트하는 데 사용됩니다.

#### 베이즈 정리의 수식:

$$
P(A \mid B) = \frac{P(B \mid A) P(A)}{P(B)}
$$

- $ P(A \mid B) $: 사건 $ B $가 발생했을 때, 사건 $ A $가 발생할 확률 (사후 확률, posterior probability).
- $ P(B \mid A) $: 사건 $ A $가 발생했을 때, 사건 $ B $가 발생할 확률 (우도, likelihood).
- $ P(A) $: 사건 $ A $가 발생할 확률 (사전 확률, prior probability).
- $ P(B) $: 사건 $ B $가 발생할 확률 (증거, evidence).

#### 예시:

어떤 질병에 걸릴 확률이 1%인 경우를 생각해봅시다. 특정 검사에서 질병이 있는 사람은 90% 확률로 양성 반응을 보이며, 질병이 없는 사람도 5% 확률로 양성 반응을 보입니다. 이때, 검사가 양성 반응을 보였을 때 실제로 질병이 있을 확률을 베이즈 정리를 통해 구할 수 있습니다.

- $ P(\text{질병}) = 0.01 $ (사전 확률)
- $ P(\text{양성} \mid \text{질병}) = 0.9 $ (우도)
- $ P(\text{양성} \mid \neg \text{질병}) = 0.05 $ (질병이 없을 때 양성일 확률)
- $ P(\text{양성}) $: 총 양성일 확률

총 양성일 확률은 전체 사건의 법칙에 따라 구할 수 있습니다:

$$
P(\text{양성}) = P(\text{양성} \mid \text{질병}) P(\text{질병}) + P(\text{양성} \mid \neg \text{질병}) P(\neg \text{질병})
$$
$$
= (0.9 \times 0.01) + (0.05 \times 0.99) = 0.009 + 0.0495 = 0.0585
$$

베이즈 정리를 적용하면:

$$
P(\text{질병} \mid \text{양성}) = \frac{P(\text{양성} \mid \text{질병}) P(\text{질병})}{P(\text{양성})}
= \frac{0.9 \times 0.01}{0.0585} \approx 0.154
$$

따라서 양성 반응을 보였을 때 실제로 질병이 있을 확률은 약 15.4%입니다.

### 3. 사건의 독립성 (Independence of Events)

사건의 독립성이란 두 사건이 서로 영향을 주지 않는다는 의미입니다. 즉, 한 사건이 일어날 확률이 다른 사건의 발생 여부에 의존하지 않는다면, 두 사건은 독립적이라고 말합니다.

#### 독립성의 정의:

두 사건 $ A $와 $ B $가 독립적일 때, 다음과 같은 관계가 성립합니다:

$$
P(A \cap B) = P(A) \times P(B)
$$

즉, 사건 $ A $와 $ B $가 독립이라면, 두 사건이 동시에 발생할 확률은 각각의 사건이 따로 발생할 확률을 곱한 것과 같습니다.

또한 독립 사건에 대해서는 조건부 확률이 다음과 같이 간단해집니다:

$$
P(A \mid B) = P(A) \quad \text{(독립적이므로 $ P(A) $는 $ B $의 발생 여부와 무관)}
$$

#### 예시:

- 동전을 두 번 던지는 실험에서, 첫 번째 던진 결과와 두 번째 던진 결과는 서로 독립적입니다. 즉, 첫 번째 던진 동전의 결과가 무엇이든 두 번째 던진 동전의 확률은 바뀌지 않습니다.

- $ A $: 첫 번째 동전이 앞면이 나옴.
- $ B $: 두 번째 동전이 앞면이 나옴.

이때, $ P(A \cap B) = P(A) \times P(B) = \frac{1}{2} \times \frac{1}{2} = \frac{1}{4} $.

두 사건 $ A $와 $ B $는 독립적입니다.

### 요약

1. 조건부 확률은 한 사건이 발생했을 때, 다른 사건이 발생할 확률을 의미하며, 사건 간의 상호 의존성을 설명합니다.
2. 베이즈 정리는 조건부 확률을 사용해 사후 확률을 계산하는 방법을 제공합니다. 이는 새로운 정보가 주어졌을 때 확률을 업데이트하는 데 유용합니다.
3. 사건의 독립성은 두 사건이 서로 영향을 미치지 않는 경우를 의미하며, 독립적 사건의 조건부 확률은 각각의 사건의 확률과 동일합니다. 

이 개념들은 확률 이론에서 매우 중요한 기초적인 도구로, 다양한 통계적 문제를 해결할 때 사용됩니다.

여러 사건의 독립성이란 두 개 이상의 사건들이 서로 영향을 주지 않는 상황을 말합니다. 이를 수학적으로 정의하려면, 모든 사건의 조합에 대해 독립성이 성립해야 합니다.

### 여러 사건의 독립성 정의

사건 $ A_1, A_2, \dots, A_n $이 서로 독립이라고 말하려면, 모든 부분집합에 대해 다음과 같은 조건이 성립해야 합니다:

$$
P(A_{i_1} \cap A_{i_2} \cap \dots \cap A_{i_k}) = P(A_{i_1}) \times P(A_{i_2}) \times \dots \times P(A_{i_k})
$$

여기서 $ \{A_{i_1}, A_{i_2}, \dots, A_{i_k}\} $는 사건 $ A_1, A_2, \dots, A_n $의 임의의 부분집합입니다.

즉, 사건 $ A_1, A_2, \dots, A_n $가 독립적이라는 말은 이들 중 임의의 사건들의 조합에 대해 모든 결합 확률이 각각의 사건 확률의 곱으로 나타나야 한다는 의미입니다.

### 독립성의 조건

1. 두 사건의 독립성:
   두 사건 $ A $와 $ B $가 독립이기 위한 조건은 다음과 같습니다:
   
   $$
   P(A \cap B) = P(A) \times P(B)
   $$
   
   이 조건은 사건 $ A $와 $ B $가 동시에 발생할 확률이 두 사건이 각각 발생할 확률의 곱과 같을 때, 두 사건이 독립임을 의미합니다.
   
2. 세 사건의 독립성:
   세 사건 $ A $, $ B $, $ C $가 독립이려면 다음 세 가지 조건이 모두 성립해야 합니다:
   
   1. 각 두 사건이 독립적이어야 합니다:
      $$
      P(A \cap B) = P(A) \times P(B)
      $$
      $$
      P(A \cap C) = P(A) \times P(C)
      $$
      $$
      P(B \cap C) = P(B) \times P(C)
      $$
   
   2. 세 사건의 결합 확률이 세 사건의 각각의 확률의 곱과 같아야 합니다:
      $$
      P(A \cap B \cap C) = P(A) \times P(B) \times P(C)
      $$
   
3. 일반적인 $ n $개의 사건의 독립성:
   $ n $개의 사건 $ A_1, A_2, \dots, A_n $이 독립이려면, 모든 부분집합에 대해 독립성이 성립해야 합니다. 즉, 각 사건들의 조합에 대한 결합 확률이 각각의 사건 확률의 곱과 같아야 합니다.

   예를 들어, 네 사건 $ A_1, A_2, A_3, A_4 $가 있을 때, 다음과 같은 조건들이 모두 성립해야 합니다:
   
   - 모든 두 사건의 독립성:
     $$
     P(A_1 \cap A_2) = P(A_1) \times P(A_2), \dots, P(A_3 \cap A_4) = P(A_3) \times P(A_4)
     $$
   
   - 모든 세 사건의 독립성:
     $$
     P(A_1 \cap A_2 \cap A_3) = P(A_1) \times P(A_2) \times P(A_3)
     $$
     $$
     \dots
     $$
   
   - 모든 네 사건의 독립성:
     $$
     P(A_1 \cap A_2 \cap A_3 \cap A_4) = P(A_1) \times P(A_2) \times P(A_3) \times P(A_4)
     $$

### 독립성의 조건 성립을 확인하는 예시

#### 예: 세 사건의 독립성

세 사건 $ A $, $ B $, $ C $가 있다고 가정해봅시다. 이들이 독립적이라는 것을 확인하려면, 다음 네 가지 조건을 확인해야 합니다:

1. $ P(A \cap B) = P(A) \times P(B) $
2. $ P(A \cap C) = P(A) \times P(C) $
3. $ P(B \cap C) = P(B) \times P(C) $
4. $ P(A \cap B \cap C) = P(A) \times P(B) \times P(C) $

이 네 조건이 모두 성립하면, 사건 $ A $, $ B $, $ C $는 서로 독립적입니다.

#### 반례: 두 사건만 독립적일 때

두 사건 $ A $와 $ B $가 독립적이더라도, 사건 $ A $, $ B $, $ C $가 모두 독립적인 것은 아닙니다. 예를 들어, $ P(A \cap B) = P(A) \times P(B) $이 성립하지만, $ P(A \cap B \cap C) \neq P(A) \times P(B) \times P(C) $일 수 있습니다. 즉, 두 사건의 독립성은 성립하더라도 세 사건 이상의 독립성은 성립하지 않을 수 있습니다.

### 독립성과 상호배타성의 차이

독립성은 사건들이 서로 영향을 미치지 않는다는 의미입니다. 즉, 한 사건이 일어나도 다른 사건이 일어날 확률이 변하지 않습니다.

반면, 상호배타성(Mutual Exclusivity)은 한 사건이 발생하면 다른 사건이 절대 발생할 수 없다는 의미입니다. 즉, 두 사건 $ A $와 $ B $가 상호배타적이라면, $ P(A \cap B) = 0 $이 됩니다. 

상호배타적 사건들은 독립적일 수 없습니다. 왜냐하면, $ P(A \cap B) = 0 $인 경우 $ P(A \times B) = P(A) \times P(B) $가 성립하지 않기 때문입니다.

### 결론

여러 사건의 독립성이 성립하려면, 모든 사건들의 조합에 대해 결합 확률이 각 사건의 확률의 곱과 같아야 합니다. 두 사건이 독립적이라고 해서 그 이상의 사건들도 반드시 독립적인 것은 아니며, 사건들의 독립성을 확인할 때는 모든 사건들의 조합에 대해 독립성이 성립하는지 확인해야 합니다.

사건 A와 B가 독립이면, A와 B의 여사건도 독립임을 증명하는 과정을 설명하겠습니다.

먼저, 독립성의 정의와 여사건의 성질을 사용하여 증명을 진행하겠습니다.

### 1. 독립성의 정의

사건 $ A $와 $ B $가 독립이라는 것은 다음과 같은 조건을 만족하는 것을 의미합니다:
$$
P(A \cap B) = P(A) \times P(B)
$$
즉, $ A $와 $ B $가 독립일 때 두 사건의 결합 확률은 각 사건의 확률의 곱과 같다는 의미입니다.

### 2. 여사건의 정의

- 사건 A의 여사건은 사건 $ A $가 발생하지 않는 경우를 의미하며, 보통 $ A^c $로 나타냅니다.
- 여사건의 확률: $ P(A^c) = 1 - P(A) $.

### 3. 증명 목표

사건 $ A $와 $ B $가 독립이라면, 사건 $ A $와 $ B $의 여사건인 $ A^c $와 $ B^c $도 독립임을 증명해야 합니다. 즉, 다음을 보여야 합니다:
$$
P(A^c \cap B^c) = P(A^c) \times P(B^c)
$$

### 4. 증명 과정

사건 $ A $와 $ B $의 여사건 $ A^c $와 $ B^c $의 교집합은 다음과 같이 표현될 수 있습니다:

$$
A^c \cap B^c = (A \cup B)^c
$$
즉, $ A^c \cap B^c $는 $ A \cup B $의 여집합입니다. 이를 기반으로 양쪽의 확률을 계산할 수 있습니다.

#### 1) $ P(A^c \cap B^c) $ 계산

여사건의 교집합 $ A^c \cap B^c $는 $ A $ 또는 $ B $ 중 어느 것도 발생하지 않는 경우입니다. 이를 여사건의 성질을 이용해 다음과 같이 쓸 수 있습니다:
$$
P(A^c \cap B^c) = 1 - P(A \cup B)
$$
그리고 덧셈 정리에 따라 $ P(A \cup B) $는 다음과 같이 표현됩니다:
$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$
$ A $와 $ B $가 독립이므로 $ P(A \cap B) = P(A) \times P(B) $를 대입하면:
$$
P(A \cup B) = P(A) + P(B) - P(A) \times P(B)
$$
따라서:
$$
P(A^c \cap B^c) = 1 - \left( P(A) + P(B) - P(A) \times P(B) \right)
$$
이 식을 정리하면:
$$
P(A^c \cap B^c) = 1 - P(A) - P(B) + P(A) \times P(B)
$$

#### 2) $ P(A^c) \times P(B^c) $ 계산

여사건의 확률을 사용하면:
$$
P(A^c) = 1 - P(A)
$$
$$
P(B^c) = 1 - P(B)
$$
따라서 $ P(A^c) \times P(B^c) $는 다음과 같이 계산됩니다:
$$
P(A^c) \times P(B^c) = (1 - P(A)) \times (1 - P(B))
$$
이를 전개하면:
$$
P(A^c) \times P(B^c) = 1 - P(A) - P(B) + P(A) \times P(B)
$$

#### 3) 두 결과의 비교

$ P(A^c \cap B^c) $와 $ P(A^c) \times P(B^c) $의 계산 결과가 모두 같음을 확인할 수 있습니다:
$$
P(A^c \cap B^c) = 1 - P(A) - P(B) + P(A) \times P(B)
$$
$$
P(A^c) \times P(B^c) = 1 - P(A) - P(B) + P(A) \times P(B)
$$

따라서, $ P(A^c \cap B^c) = P(A^c) \times P(B^c) $가 성립합니다.

### 5. 결론

사건 $ A $와 $ B $가 독립이라면, 그 여사건 $ A^c $와 $ B^c $도 독립임을 증명하였습니다. 이는 여사건의 결합 확률이 각 여사건의 확률의 곱과 같음을 보임으로써 증명되었습니다.

사건 $A$와 사건 $B$가 독립이라면, 사건 $A$와 사건 $B$의 여사건 $B^c$도 독립인 경우를 증명하는 과정을 설명하겠습니다.

### 목표

사건 $A$와 사건 $B$가 독립이라면, $A$와 $B^c$도 독립임을 보이는 것이 목표입니다. 즉, 우리는 아래의 관계를 증명해야 합니다:

$$
P(A \cap B^c) = P(A) \times P(B^c)
$$

### 독립성의 정의

사건 $A$와 사건 $B$가 독립이라는 것은 다음을 의미합니다:
$$
P(A \cap B) = P(A) \times P(B)
$$

### 1. $ P(A \cap B^c) $의 계산

사건 $A$와 사건 $B^c$의 교집합 $ P(A \cap B^c) $를 계산하는데, 이를 여사건의 성질을 이용해 표현할 수 있습니다. 

사건 $B^c$는 사건 $B$의 여사건이므로, 다음과 같이 쓸 수 있습니다:

$$
P(A \cap B^c) = P(A) - P(A \cap B)
$$

이 식은 전체 사건 $A$에서 사건 $A$와 $B$가 동시에 일어날 확률 $P(A \cap B)$를 빼면, $A$와 $B^c$가 동시에 일어나는 확률이 남는다는 것을 의미합니다.

### 2. $ P(A \cap B^c) $ 계산에 독립성 사용

사건 $A$와 $B$가 독립이므로, $ P(A \cap B) = P(A) \times P(B) $입니다. 이를 위의 식에 대입하면:

$$
P(A \cap B^c) = P(A) - P(A) \times P(B)
$$

### 3. $ P(B^c) $ 계산

사건 $B^c$의 확률은 여사건의 성질에 따라 다음과 같이 계산됩니다:

$$
P(B^c) = 1 - P(B)
$$

### 4. $ P(A) \times P(B^c) $ 계산

다음으로, $ P(A) \times P(B^c) $를 계산해 보겠습니다:

$$
P(A) \times P(B^c) = P(A) \times (1 - P(B)) = P(A) - P(A) \times P(B)
$$

### 5. 두 결과 비교

우리가 구한 두 개의 확률:

- $ P(A \cap B^c) = P(A) - P(A) \times P(B) $
- $ P(A) \times P(B^c) = P(A) - P(A) \times P(B) $

이 두 값이 동일함을 확인할 수 있습니다.

### 6. 결론

따라서, 사건 $A$와 $B$가 독립이라면, 사건 $A$와 사건 $B^c$ (사건 $B$의 여사건)도 독립임을 증명할 수 있습니다.

즉, $ P(A \cap B^c) = P(A) \times P(B^c) $가 성립하므로, 사건 $A$와 $B^c$는 독립입니다.

# 확률변수와 확률분포

확률변수(Probability Variable)는 확률론과 통계학에서 매우 중요한 개념으로, 무작위 실험의 결과를 수치적으로 표현하는 함수입니다. 확률변수는 실험의 결과를 수치로 매핑하여, 우리가 실험에서 기대할 수 있는 값들을 분석하는 데 사용됩니다.

### 확률변수의 정의

확률변수는 무작위 실험에서 발생할 수 있는 모든 결과를 수치로 대응시키는 함수입니다. 즉, 확률변수는 표본 공간에서 얻은 결과를 실수 집합에 대응시키는 역할을 합니다.

- 표본 공간(Sample Space, $S$): 확률 실험에서 가능한 모든 결과들의 집합.
- 확률변수(Random Variable, $X$): 표본 공간의 원소를 실수로 변환하는 함수.

#### 예시:
동전을 두 번 던지는 실험을 생각해봅시다. 이때 표본 공간 $ S $는 다음과 같습니다:
$$
S = \{(HH), (HT), (TH), (TT)\}
$$

확률변수 $ X $를 "앞면이 나온 횟수"로 정의할 수 있습니다. 이 경우 확률변수는 아래와 같이 표본 공간의 각 원소를 실수로 대응시킵니다:
- $ X(HH) = 2 $
- $ X(HT) = 1 $
- $ X(TH) = 1 $
- $ X(TT) = 0 $

### 1. 이산형 확률변수 (Discrete Random Variable)

이산형 확률변수는 특정한 개수의 구별되는 값을 가지는 확률변수입니다. 주사위를 던지는 실험에서 주사위 눈의 결과와 같이 값들이 이산적(카운팅 가능한 값)일 때, 이를 이산형 확률변수라고 합니다.

#### 이산형 확률변수의 확률 분포:
이산형 확률변수의 확률 분포는 확률 질량 함수(Probability Mass Function, PMF)로 나타내며, 이는 특정 값에 대응하는 확률을 정의합니다.

- 이산형 확률변수 $ X $가 가질 수 있는 값들이 $ x_1, x_2, \dots $일 때, 확률 질량 함수 $ p(x_i) = P(X = x_i) $는 해당 값이 나올 확률을 나타냅니다.
- 모든 이산형 확률변수는 다음과 같은 성질을 만족해야 합니다:
  $$
  \sum_{i} P(X = x_i) = 1
  $$

#### 예시:
주사위를 던지는 실험에서 확률변수 $ X $를 주사위 눈의 값으로 정의한다면, 확률 질량 함수는 다음과 같이 정의됩니다:
$$
P(X = 1) = P(X = 2) = P(X = 3) = P(X = 4) = P(X = 5) = P(X = 6) = \frac{1}{6}
$$

### 2. 연속형 확률변수 (Continuous Random Variable)

연속형 확률변수는 연속적인 값을 가질 수 있는 확률변수입니다. 연속형 확률변수는 구간 내에서 무한히 많은 값을 가질 수 있으며, 특정 값이 아닌 구간에 대해 확률을 계산합니다.

#### 연속형 확률변수의 확률 분포:
연속형 확률변수의 확률 분포는 확률 밀도 함수(Probability Density Function, PDF)로 나타냅니다. 특정 값에서의 확률은 항상 0이므로, 구간에 대한 확률을 계산합니다.

- $ f_X(x) $는 확률 밀도 함수로, 확률변수 $ X $의 값이 주어진 구간에서 얼마나 분포되는지를 나타냅니다.
- 연속형 확률변수는 다음과 같은 성질을 만족합니다:
  $$
  P(a \leq X \leq b) = \int_a^b f_X(x) dx
  $$
  즉, $ X $가 구간 $ [a, b] $에 속할 확률은 해당 구간에서의 확률 밀도 함수의 적분으로 계산됩니다.

#### 예시:
하루 동안의 기온은 연속적인 값입니다. 이때 확률변수 $ X $를 "하루 동안의 기온"으로 정의하면, 확률 밀도 함수를 통해 특정 범위의 온도가 발생할 확률을 계산할 수 있습니다.

### 3. 확률분포 (Probability Distribution)

확률변수는 확률분포(Probability Distribution)를 따릅니다. 확률분포는 확률변수가 취할 수 있는 값과 그 값이 나타날 확률을 정의하는 함수입니다.

#### 이산형 확률분포:
이산형 확률변수의 경우, 확률 질량 함수(PMF)를 통해 확률분포를 나타냅니다. 대표적인 이산형 확률분포로는 이항분포와 포아송분포가 있습니다.

- 이항분포: 성공 또는 실패가 가능한 독립적인 실험을 여러 번 반복할 때 나타나는 분포입니다.
- 포아송분포: 주어진 시간 동안 특정 사건이 발생하는 횟수를 모델링할 때 사용됩니다.

#### 연속형 확률분포:
연속형 확률변수의 경우, 확률 밀도 함수(PDF)를 통해 확률분포를 나타냅니다. 대표적인 연속형 확률분포로는 정규분포, 균등분포, 지수분포 등이 있습니다.

- 정규분포: 평균을 중심으로 대칭적인 종 모양의 분포를 가집니다. 자연 현상에서 많이 나타나는 분포입니다.
- 균등분포: 특정 구간 내에서 모든 값이 동일한 확률로 발생하는 분포입니다.
- 지수분포: 사건 간의 시간 간격을 모델링하는 데 사용됩니다.

### 4. 기대값과 분산

확률변수의 중요한 특성으로 기대값(Expectation)과 분산(Variance)이 있습니다.

#### 1) 기대값
확률변수의 기대값은 확률변수가 가질 수 있는 값들의 평균을 나타냅니다.

- 이산형 확률변수의 기대값:
  $$
  E(X) = \sum_{i} x_i P(X = x_i)
  $$
  
- 연속형 확률변수의 기대값:
  $$
  E(X) = \int_{-\infty}^{\infty} x f_X(x) dx
  $$

#### 2) 분산
확률변수의 분산은 값들이 기대값을 중심으로 얼마나 퍼져 있는지를 나타냅니다.

- 이산형 확률변수의 분산:
  $$
  \text{Var}(X) = \sum_{i} (x_i - E(X))^2 P(X = x_i)
  $$

- 연속형 확률변수의 분산:
  $$
  \text{Var}(X) = \int_{-\infty}^{\infty} (x - E(X))^2 f_X(x) dx
  $$

### 5. 결합 확률변수 (Joint Random Variables)

두 개 이상의 확률변수들이 있을 때, 이들의 관계를 나타내는 결합 확률변수(Joint Random Variables) 개념이 있습니다. 예를 들어, 두 확률변수 $ X $와 $ Y $가 동시에 가질 수 있는 값을 다룹니다.

#### 결합 확률 분포:
- 이산형 확률변수의 결합 확률분포는 $ P(X = x, Y = y) $로 나타냅니다.
- 연속형 확률변수의 결합 확률밀도 함수는 $ f_{X,Y}(x, y) $로 나타냅니다.

### 요약

- 확률변수는 확률 실험에서 얻은 결과를 수치적으로 나타내는 함수입니다.
- 확률변수는 이산형과 연속형으로 나뉘며, 각 유형에 따라 확률 질량 함수(PMF) 또는 확률 밀도 함수(PDF)로 표현됩니다.
- 확률변수는 확률분포를 따르며, 확률변수의 특성으로 기대값과 분산이 중요합니다.
- 여러 확률변수 간의 관계는 결합 확률변수로 나타낼 수 있습니다.

이 개념들을 통해 무작위 실험의 결과를 수치적으로 분석하고, 확률론적 문제를 해결할 수 있습니다.

지표함수(Indicator Function)는 수학, 확률론, 통계학에서 특정 사건이 발생했는지 여부를 수치적으로 나타내는 함수입니다. 지표함수는 주어진 집합에서 어떤 조건이 성립하는지 여부를 1과 0으로 표시하는 함수로, 어떤 사건이 발생하면 1을, 그렇지 않으면 0을 반환합니다.

지표함수는 다음과 같이 정의됩니다:

### 지표함수의 정의

어떤 사건 $ A $에 대해, 지표함수 $ I_A(x) $는 다음과 같이 정의됩니다:
$$
I_A(x) = \begin{cases}
1 & \text{if } x \in A \\
0 & \text{if } x \notin A
\end{cases}
$$
여기서:
- $ A $: 특정 조건을 만족하는 사건이나 집합.
- $ x $: 실험이나 상황에서 고려되는 값.

즉, 지표함수 $ I_A(x) $는 값 $ x $가 사건 $ A $에 포함되는지 여부를 1 또는 0으로 나타냅니다.

### 지표함수의 예

#### 예시 1: 주사위 예시

주사위를 던지는 실험에서, 사건 $ A $를 "나온 숫자가 4 이상"이라고 정의하겠습니다. 주사위가 던져질 때, 지표함수 $ I_A(x) $는 주사위 눈의 결과가 4 이상이면 1을, 그렇지 않으면 0을 반환합니다.

- $ A = \{4, 5, 6\} $
- $ I_A(3) = 0 $ (3은 4 이상이 아니므로)
- $ I_A(5) = 1 $ (5는 4 이상이므로)

#### 예시 2: 확률 변수와 지표함수

어떤 확률 변수 $ X $가 주어졌을 때, 지표함수 $ I_{\{X \geq 2\}} $는 $ X $가 2 이상일 때 1, 그렇지 않으면 0을 반환합니다. 예를 들어, $ X = 3 $이면 $ I_{\{X \geq 2\}} = 1 $, $ X = 1 $이면 $ I_{\{X \geq 2\}} = 0 $입니다.

### 지표함수의 성질

1. 덧셈 성질: 두 사건 $ A $와 $ B $에 대해, $ A \cup B $의 지표함수는 두 사건의 지표함수의 합에서 교집합의 지표함수를 빼는 것과 같습니다:
   $$
   I_{A \cup B}(x) = I_A(x) + I_B(x) - I_{A \cap B}(x)
   $$
   
2. 곱셈 성질: 두 사건 $ A $와 $ B $에 대해, $ A \cap B $의 지표함수는 두 사건의 지표함수의 곱과 같습니다:
   $$
   I_{A \cap B}(x) = I_A(x) \cdot I_B(x)
   $$

3. 여사건과 지표함수: 사건 $ A $의 여사건 $ A^c $에 대한 지표함수는 다음과 같은 관계를 가집니다:
   $$
   I_{A^c}(x) = 1 - I_A(x)
   $$

### 지표함수의 활용

1. 통계적 계산:
   - 지표함수는 특정 사건의 발생 여부를 수치로 나타내므로, 이를 활용하여 평균이나 확률을 계산할 수 있습니다. 예를 들어, 사건 $ A $가 발생할 확률 $ P(A) $는 지표함수의 기대값으로 표현할 수 있습니다:
     $$
     P(A) = E[I_A(X)]
     $$
     여기서 $ E $는 기대값을 나타냅니다.

2. 구간에 대한 계산:
   - 지표함수는 구간 내에서 조건을 정의할 때 유용합니다. 예를 들어, 어떤 변수가 특정 구간 내에 있는지 확인할 때 지표함수를 사용하여 조건을 간단하게 표현할 수 있습니다:
     $$
     I_{\{a \leq X \leq b\}}(x) = \begin{cases}
     1 & \text{if } a \leq x \leq b \\
     0 & \text{otherwise}
     \end{cases}
     $$

3. 확률과 통계에서의 응용:
   - 지표함수는 이항분포, 포아송분포 등에서 특정 사건의 발생 횟수를 계산하는 데 자주 사용됩니다.
   - 또한, 대형 샘플의 개별 데이터가 특정 조건을 만족하는지를 수학적으로 표현할 때 사용되기도 합니다.

### 요약

- 지표함수(Indicator Function)는 특정 사건이 발생했는지 여부를 나타내는 함수로, 사건이 발생하면 1, 그렇지 않으면 0을 반환합니다.
- 주로 확률, 통계, 집합론 등에서 사건의 발생 여부를 수치적으로 표현하는 데 사용됩니다.
- 지표함수는 특정 조건을 만족하는지 여부를 표현하는 간단하고 유용한 도구로서, 확률론적 계산과 통계적 분석에서 자주 사용됩니다.

# 확률분포의 기대값

확률분포의 기대값(Expectation of a Probability Distribution)은 확률변수가 가질 수 있는 값들의 가중평균을 의미하며, 확률분포의 중심 경향을 나타냅니다. 기대값은 확률변수가 반복적으로 실현될 때 그 값들의 평균을 의미합니다.

확률분포의 기대값은 이산형 확률변수와 연속형 확률변수에 대해 다르게 정의됩니다. 각 경우에 대해 설명하겠습니다.

### 1. 이산형 확률변수의 기대값

이산형 확률변수의 기대값은 각 값이 발생할 확률을 그 값에 곱한 후, 모든 값에 대해 합한 값입니다. 즉, 확률 질량 함수(PMF)를 사용하여 기대값을 계산합니다.

#### 정의:
이산형 확률변수 $ X $가 가질 수 있는 값이 $ x_1, x_2, \dots $이고, 그 값들이 각각 $ P(X = x_i) = p_i $의 확률로 발생한다고 하면, 기대값 $ E(X) $는 다음과 같이 정의됩니다:

$$
E(X) = \sum_{i} x_i \cdot P(X = x_i)
$$

즉, 확률변수 $ X $가 가질 수 있는 모든 값 $ x_i $에 대해, 그 값에 해당하는 확률을 곱한 후 모두 더한 값이 기대값입니다.

#### 예시:
주사위를 던지는 실험에서, 확률변수 $ X $를 주사위 눈의 값으로 정의하면:
- $ X $의 값은 1, 2, 3, 4, 5, 6이고,
- 각 값이 나올 확률은 $ P(X = x_i) = \frac{1}{6} $입니다.

이때 기대값 $ E(X) $는 다음과 같이 계산됩니다:
$$
E(X) = 1 \cdot \frac{1}{6} + 2 \cdot \frac{1}{6} + 3 \cdot \frac{1}{6} + 4 \cdot \frac{1}{6} + 5 \cdot \frac{1}{6} + 6 \cdot \frac{1}{6}
$$
$$
E(X) = \frac{1 + 2 + 3 + 4 + 5 + 6}{6} = \frac{21}{6} = 3.5
$$

따라서, 주사위를 여러 번 던졌을 때 눈의 평균값은 3.5로 수렴합니다.

### 2. 연속형 확률변수의 기대값

연속형 확률변수의 기대값은 확률 밀도 함수(PDF)를 사용하여 계산합니다. 연속형 확률변수에서는 특정 값에 대한 확률은 0이므로, 특정 구간에 걸쳐 확률변수가 취할 수 있는 값들의 확률을 고려해야 합니다.

#### 정의:
연속형 확률변수 $ X $의 확률 밀도 함수가 $ f_X(x) $로 주어졌을 때, $ X $의 기대값 $ E(X) $는 다음과 같이 적분으로 정의됩니다:

$$
E(X) = \int_{-\infty}^{\infty} x \cdot f_X(x) \, dx
$$

즉, $ x $와 그에 대한 확률밀도 함수의 곱을 적분하여 기대값을 계산합니다.

#### 예시:
정규분포 $ N(\mu, \sigma^2) $의 확률밀도함수는 다음과 같이 주어집니다:
$$
f_X(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}}
$$
이때, 정규분포의 기대값 $ E(X) $는 정규분포의 평균에 해당하며, 이는 바로 $ \mu $입니다.

따라서, $ N(\mu, \sigma^2) $ 분포를 따르는 확률변수 $ X $의 기대값은:
$$
E(X) = \mu
$$

이는 정규분포의 평균이 곧 기대값임을 의미합니다.

### 3. 기대값의 성질

확률변수의 기대값은 몇 가지 중요한 성질을 가지고 있습니다.

1. 선형성:
   - 두 확률변수 $ X $와 $ Y $에 대해, 기대값은 다음과 같은 선형성을 가집니다:
   $$
   E(aX + bY) = aE(X) + bE(Y)
   $$
   여기서 $ a $와 $ b $는 상수입니다.
   
2. 상수의 기대값:
   - 상수 $ c $에 대해, 기대값은 상수 그 자체입니다:
   $$
   E(c) = c
   $$

3. 독립 확률변수의 기대값:
   - 두 확률변수 $ X $와 $ Y $가 독립이면, $ X \cdot Y $의 기대값은 각각의 기대값의 곱과 같습니다:
   $$
   E(X \cdot Y) = E(X) \cdot E(Y)
   $$

### 4. 분산과 기대값의 관계

분산(Variance)은 확률변수가 기대값 주변에서 얼마나 퍼져 있는지를 나타내는 값입니다. 확률변수 $ X $의 분산 $ \text{Var}(X) $는 다음과 같이 정의됩니다:

$$
\text{Var}(X) = E\left[(X - E(X))^2\right]
$$

이는 확률변수가 기대값으로부터 얼마나 떨어져 있는지의 평균 제곱 거리를 의미합니다.

#### 분산의 또 다른 표현:
분산은 기대값을 사용하여 다음과 같이 다시 표현할 수 있습니다:
$$
\text{Var}(X) = E(X^2) - (E(X))^2
$$

이 식은 확률변수의 제곱된 값의 기대값에서, 기대값의 제곱을 뺀 값으로 분산을 구하는 방법을 제시합니다.

### 5. 응용

확률변수의 기대값은 통계학 및 확률론에서 매우 중요한 개념으로, 다양한 분야에서 응용됩니다. 기대값을 통해 데이터의 중심 경향을 파악하거나, 확률변수의 평균적인 행동을 예측할 수 있습니다. 예를 들어:

- 보험 수학: 보험료 산정에서 기대 손실을 계산하여 적정한 보험료를 책정합니다.
- 재무학: 주식이나 투자 상품의 평균 수익을 기대값으로 계산하여 투자 결정을 내립니다.
- 게임 이론: 게임에서 발생할 수 있는 평균적인 보상을 계산하여 전략을 세우는 데 사용됩니다.

### 요약

- 기대값은 확률변수가 취할 수 있는 값들의 평균을 나타내는 개념입니다.
- 이산형 확률변수는 확률 질량 함수(PMF)를 사용하여 기대값을 계산하고, 연속형 확률변수는 확률 밀도 함수(PDF)를 사용하여 기대값을 적분으로 계산합니다.
- 기대값은 선형성을 가지고 있으며, 분산은 기대값을 기반으로 확률변수가 얼마나 퍼져 있는지를 나타냅니다.
- 기대값은 다양한 분야에서 데이터의 중심 경향을 분석하는 중요한 도구로 사용됩니다.

# 누적분포함수

누적분포함수(Cumulative Distribution Function, CDF)는 확률변수 $X$가 특정 값 이하일 확률을 나타내는 함수입니다. 다시 말해, 누적분포함수는 확률변수가 주어진 값보다 작거나 같은 값을 가질 확률을 계산하는 데 사용됩니다.

### 1. 누적분포함수의 정의

확률변수 $X$의 누적분포함수 $F_X(x)$는 다음과 같이 정의됩니다:

$$
F_X(x) = P(X \leq x)
$$

즉, 누적분포함수 $F_X(x)$는 확률변수 $X$가 주어진 값 $x$보다 작거나 같은 확률을 의미합니다. 이는 이산형 또는 연속형 확률변수 모두에 대해 적용됩니다.

### 2. 이산형 확률변수의 누적분포함수

이산형 확률변수 $X$에 대해 누적분포함수는 각 가능한 값까지의 확률의 합으로 정의됩니다. 이산형 확률변수의 경우 특정 값에서 확률이 비연속적으로 점프하게 됩니다.

#### 예시:
주사위를 던지는 실험에서, 확률변수 $X$를 주사위 눈의 값으로 정의하면 $X$는 이산형 확률변수입니다. 이때 $X$의 누적분포함수 $F_X(x)$는 다음과 같이 계산됩니다:
$$
F_X(x) = P(X \leq x) = \sum_{k \leq x} P(X = k)
$$
이를 각 값에 대해 계산해 보면:
- $ F_X(1) = P(X \leq 1) = \frac{1}{6} $
- $ F_X(2) = P(X \leq 2) = \frac{2}{6} $
- $ F_X(3) = P(X \leq 3) = \frac{3}{6} $
- $ F_X(4) = P(X \leq 4) = \frac{4}{6} $
- $ F_X(5) = P(X \leq 5) = \frac{5}{6} $
- $ F_X(6) = P(X \leq 6) = 1 $

### 3. 연속형 확률변수의 누적분포함수

연속형 확률변수의 경우, 누적분포함수는 확률 밀도 함수(PDF)의 적분으로 정의됩니다. 즉, 특정 값 $x$까지의 구간에 대해 확률 밀도 함수의 값을 적분하여 누적분포함수를 계산합니다.

#### 정의:
연속형 확률변수 $X$의 확률 밀도 함수가 $f_X(x)$일 때, 누적분포함수 $F_X(x)$는 다음과 같이 정의됩니다:

$$
F_X(x) = P(X \leq x) = \int_{-\infty}^{x} f_X(t) \, dt
$$

즉, $X$가 $x$보다 작거나 같은 확률을 구하기 위해, 확률 밀도 함수를 $ -\infty $에서 $ x $까지 적분합니다.

#### 예시:
정규분포 $ N(\mu, \sigma^2) $를 따르는 연속형 확률변수 $X$에 대해, 그 확률 밀도 함수는 다음과 같습니다:

$$
f_X(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

이때, 누적분포함수 $F_X(x)$는 다음과 같이 적분으로 계산됩니다:

$$
F_X(x) = \int_{-\infty}^{x} \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(t-\mu)^2}{2\sigma^2}} \, dt
$$

정규분포의 누적분포함수는 기본적으로 오차 함수(erf)를 사용해 계산되며, 이를 통해 정규분포에서 특정 값까지의 확률을 구할 수 있습니다.

### 4. 누적분포함수의 성질

누적분포함수는 다음과 같은 중요한 성질을 가집니다:

1. 비감소성:
   $$
   F_X(x_1) \leq F_X(x_2) \quad \text{for all } x_1 \leq x_2
   $$
   즉, 누적분포함수는 $ x $가 증가할수록 감소하지 않습니다.

2. 경계 값:
   - $ \lim_{x \to -\infty} F_X(x) = 0 $ (확률변수가 아주 작은 값일 확률은 0에 수렴).
   - $ \lim_{x \to \infty} F_X(x) = 1 $ (확률변수가 아주 큰 값일 확률은 1에 수렴).

3. 확률 계산:
   - 누적분포함수를 사용해 구간 확률을 계산할 수 있습니다. 예를 들어, 확률변수 $ X $가 구간 $ (a, b) $에 있을 확률은 다음과 같이 계산됩니다:
     $$
     P(a < X \leq b) = F_X(b) - F_X(a)
     $$

4. 연속형 확률변수의 도함수:
   - 연속형 확률변수의 경우, 누적분포함수의 도함수는 확률 밀도 함수(PDF)에 해당합니다:
     $$
     f_X(x) = \frac{d}{dx} F_X(x)
     $$
     즉, PDF는 CDF의 미분입니다.

### 5. 누적분포함수와 기대값

누적분포함수는 확률변수의 기대값을 계산하는 데에도 중요한 역할을 합니다. 기대값은 확률변수의 값에 대한 확률의 가중 평균이므로, 누적분포함수를 통해 구간 확률을 계산함으로써 기대값을 구할 수 있습니다.

---

### 요약

- 누적분포함수(CDF)는 확률변수가 주어진 값보다 작거나 같을 확률을 나타내는 함수입니다.
- 이산형 확률변수에서는 특정 값까지의 확률의 합으로, 연속형 확률변수에서는 확률 밀도 함수의 적분으로 계산됩니다.
- 누적분포함수는 비감소 함수이며, 특정 구간 내에서 확률변수가 존재할 확률을 계산하는 데 유용합니다.
- 연속형 확률변수의 경우, 누적분포함수의 미분은 확률 밀도 함수(PDF)와 동일합니다.

누적분포함수는 확률이 특정 값 이하에서 누적되는 과정을 나타내기 때문에, 확률론과 통계학에서 중요한 역할을 합니다.

# 생성함수

이미지에서 나오는 내용은 생성함수와 관련된 확률론의 중요한 개념을 설명하고 있습니다. 주요 생성함수는 확률생성함수(PGF), 적률생성함수(MGF), 그리고 누율생성함수(CGF)입니다. 각 생성함수는 확률분포의 특성을 분석하고, 기대값, 분산, 적률 등을 계산하는 데 유용한 도구입니다. 아래에서 각 생성함수에 대한 정의와 성질을 설명하겠습니다.

### 1. 확률생성함수 (Probability Generating Function, PGF)

확률생성함수는 이산형 확률변수의 분포를 나타내는 데 사용되는 함수입니다. 주로 이산형 확률변수에서 활용되며, 확률변수의 적률(기대값, 분산 등)을 계산하는 데 유용합니다.

- 확률생성함수 $ G(s) $는 다음과 같이 정의됩니다:
  $$
  G(s) = E(s^X)
  $$
  여기서 $ X $는 이산형 확률변수, $ E $는 기대값을 나타내며, $ s $는 보조 변수입니다. 이 함수는 확률변수 $ X $의 분포를 고유하게 나타내며, 이를 통해 확률질량함수(PMF)에서 필요한 정보를 얻을 수 있습니다.

#### 예시:
주사위를 던지는 실험에서 확률변수 $ X $를 주사위의 값이라고 하면, $ G(s) $는 각 $ X $의 값에 해당하는 확률과 $ s^X $의 합으로 계산됩니다.

### 2. 적률생성함수 (Moment Generating Function, MGF)

적률생성함수는 확률변수의 적률(즉, 기대값, 분산, 고차 적률)을 계산하는 데 사용됩니다. 적률생성함수는 확률분포의 정보를 한 함수로 요약하여 나타내며, 주어진 확률변수의 모든 적률을 계산할 수 있는 도구입니다.

- 적률생성함수 $ M(t) $는 다음과 같이 정의됩니다:
  $$
  M(t) = E(e^{tX})
  $$
  여기서 $ t $는 보조 변수이고, $ X $는 확률변수입니다. $ M(t) $는 확률변수의 모든 적률을 포함하는 함수입니다. 즉, $ M(t) $의 $ t $에 대한 도함수를 적절히 계산하면 각 차수의 적률(예: 기대값, 분산, 3차 적률 등)을 얻을 수 있습니다.

#### 적률생성함수의 성질:
- $ M(t) $의 도함수들을 이용해 기대값, 분산, 고차 적률을 계산할 수 있습니다:
  - 기대값: $ E(X) = M'(0) $
  - 분산: $ \text{Var}(X) = M''(0) - (M'(0))^2 $
  
  이러한 성질은 확률분포의 중심 경향과 변동성을 계산하는 데 매우 유용합니다.

### 3. 누율생성함수 (Cumulant Generating Function, CGF)

누율생성함수(CGF)는 적률생성함수의 로그로 정의되며, 확률분포의 누적적인 특성을 분석하는 데 사용됩니다. CGF는 확률분포의 누적 적률을 계산할 수 있는 도구로, 고차 적률과 관련된 중요한 정보를 제공합니다.

- 누율생성함수 $ C(t) $는 적률생성함수의 로그로 정의됩니다:
  $$
  C(t) = \log M(t)
  $$
  
#### 누율생성함수의 성질:
- CGF는 적률보다 더 간단하게 고차 적률을 분석할 수 있는 도구로 사용됩니다. $ C(t) $의 도함수는 누적량(cumulants)을 제공합니다. 누적량은 확률분포의 중심 경향 및 고차 변동성을 파악하는 데 사용됩니다.
- CGF의 첫 번째, 두 번째 도함수는 각각 평균(기대값)과 분산을 계산할 수 있습니다.
  - 첫 번째 도함수: $ C'(0) = E(X) $
  - 두 번째 도함수: $ C''(0) = \text{Var}(X) $

### 요약

1. 확률생성함수(PGF)는 이산형 확률변수의 분포를 나타내는 함수로, 각 값의 확률을 이용해 분포의 특성을 분석하는 도구입니다.
   
2. 적률생성함수(MGF)는 확률변수의 적률을 계산하는 데 사용되며, 기대값과 분산을 포함한 확률변수의 중요한 특성들을 계산할 수 있습니다.
   
3. 누율생성함수(CGF)는 적률생성함수의 로그로 정의되며, 확률분포의 누적 특성을 분석하는 데 사용됩니다. 이는 분포의 중심 경향과 고차 변동성을 설명하는 데 유용한 도구입니다.

각 생성함수는 확률변수의 특성을 계산하고 분석하는 데 중요한 역할을 하며, 확률 및 통계에서 매우 유용하게 사용됩니다.

# 중요한 정리들

마코프 부등식은 확률변수가 주어진 값 이상일 확률에 대한 상한을 제공하는 중요한 부등식입니다. 이 부등식은 확률분포에 대한 정보가 없더라도 확률변수가 기대값에 따라 얼마나 큰 값일지에 대한 상한을 제공합니다. 이제 마코프 부등식의 증명 과정을 살펴보겠습니다.

### 마코프 부등식의 정리

비음수 확률변수 $ X $와 양수 $ a $에 대해, 마코프 부등식은 다음과 같습니다:

$$
P(X \geq a) \leq \frac{E(X)}{a}
$$

여기서:
- $ P(X \geq a) $: 확률변수 $ X $가 $ a $ 이상일 확률.
- $ E(X) $: 확률변수 $ X $의 기대값.
- $ a $: 양의 상수.

### 증명 과정

#### 1. 지표 함수 사용

먼저 $ X \geq a $라는 조건을 만족하는 경우와 그렇지 않은 경우를 분리해보겠습니다. 이를 위해 지표 함수(Indicator function)를 사용합니다.

$$
I_{\{X \geq a\}} =
\begin{cases}
1, & \text{if } X \geq a, \\
0, & \text{if } X < a.
\end{cases}
$$

이 지표 함수 $ I_{\{X \geq a\}} $는 $ X $가 $ a $ 이상일 때 1을, 그렇지 않으면 0을 반환하는 함수입니다.

따라서, $ P(X \geq a) $는 이 지표 함수의 기대값과 같게 나타낼 수 있습니다:
$$
P(X \geq a) = E(I_{\{X \geq a\}})
$$

#### 2. 기대값에 대한 상한 설정

이제 확률변수 $ X $의 값을 지표 함수와 결합하여 상한을 설정합니다. $ X $가 $ a $ 이상인 경우 $ X \geq a $이므로, 다음 부등식이 성립합니다:
$$
X \geq a \cdot I_{\{X \geq a\}}
$$

여기서 $ X $는 $ a $ 이상일 때 지표 함수가 1이 되기 때문에, 전체 부등식은 성립합니다. 이 부등식 양변에 기대값 $ E $를 취하면:

$$
E(X) \geq E(a \cdot I_{\{X \geq a\}})
$$

기대값의 선형성에 의해 우변은 다음과 같이 정리할 수 있습니다:

$$
E(X) \geq a \cdot E(I_{\{X \geq a\}})
$$

하지만 $ E(I_{\{X \geq a\}}) $는 앞서 정의한 바와 같이 $ P(X \geq a) $이므로, 이를 대입하면:

$$
E(X) \geq a \cdot P(X \geq a)
$$

#### 3. 마코프 부등식 도출

이제 양변을 $ a $로 나누면 마코프 부등식을 얻을 수 있습니다:

$$
P(X \geq a) \leq \frac{E(X)}{a}
$$

따라서, 마코프 부등식이 성립함을 증명할 수 있습니다.

### 결론

마코프 부등식은 확률변수가 특정 값 이상이 될 확률을 그 확률변수의 기대값을 기반으로 상한을 제공합니다. 이 부등식은 비음수 확률변수에 대해 항상 성립하며, 기대값을 이용하여 확률변수의 분포에 대한 구체적인 정보가 없더라도 확률의 상한을 설정할 수 있습니다.

$$
P(X \geq a) \leq \frac{E(X)}{a}
$$

이는 확률변수가 큰 값을 가질 확률이 기대값에 반비례함을 보여줍니다.

체비셰프 부등식(Chebyshev's Inequality)은 확률변수가 기대값에서 벗어날 확률을 분산을 기반으로 상한을 제공합니다. 이 부등식은 분포의 모양에 관계없이 적용될 수 있습니다. 체비셰프 부등식의 증명 과정을 살펴보겠습니다.

### 체비셰프 부등식의 정리

확률변수 $ X $에 대해 기대값 $ \mu = E(X) $와 분산 $ \sigma^2 = \text{Var}(X) $이 있을 때, 임의의 양수 $ k $에 대해 체비셰프 부등식은 다음과 같이 표현됩니다:

$$
P(|X - \mu| \geq k\sigma) \leq \frac{1}{k^2}
$$

즉, 확률변수 $ X $가 기대값 $ \mu $에서 $ k $배 이상의 표준편차 $ \sigma $만큼 벗어날 확률이 $ \frac{1}{k^2} $로 제한된다는 의미입니다.

### 증명 과정

#### 1. 체비셰프 부등식의 형태 재구성

먼저 체비셰프 부등식에서 고려하고자 하는 확률은 $ P(|X - \mu| \geq k\sigma) $입니다. 이를 절대값을 풀어 다음과 같이 다시 쓸 수 있습니다:

$$
P(|X - \mu| \geq k\sigma) = P((X - \mu)^2 \geq k^2 \sigma^2)
$$

이제, 우리는 $ (X - \mu)^2 $이 $ k^2 \sigma^2 $ 이상일 확률에 대해 상한을 구하고자 합니다.

#### 2. 마코프 부등식 사용

확률변수 $ Y = (X - \mu)^2 $로 새로운 확률변수를 정의하겠습니다. 이제 $ Y $는 $ (X - \mu)^2 $이며, 이는 비음수 확률변수입니다. 이 변수에 대해 마코프 부등식을 적용할 수 있습니다. 마코프 부등식에 따르면:

$$
P(Y \geq a) \leq \frac{E(Y)}{a}
$$

여기서 $ a = k^2 \sigma^2 $이므로, 다음과 같은 부등식을 얻습니다:

$$
P((X - \mu)^2 \geq k^2 \sigma^2) \leq \frac{E((X - \mu)^2)}{k^2 \sigma^2}
$$

#### 3. 분산의 성질 적용

확률변수 $ X $에 대해 $ E((X - \mu)^2) $는 분산 $ \sigma^2 $와 같으므로, 이를 대입하면:

$$
P((X - \mu)^2 \geq k^2 \sigma^2) \leq \frac{\sigma^2}{k^2 \sigma^2}
$$

이를 정리하면:

$$
P(|X - \mu| \geq k\sigma) \leq \frac{1}{k^2}
$$

#### 4. 결론

따라서, 체비셰프 부등식이 성립함을 증명할 수 있습니다. 이 부등식은 확률변수가 기대값에서 크게 벗어나는 경우의 확률이 분산에 의해 제한된다는 중요한 사실을 제공합니다.

$$
P(|X - \mu| \geq k\sigma) \leq \frac{1}{k^2}
$$

이 부등식은 확률변수가 기대값에서 벗어날 확률을 분산을 바탕으로 상한할 수 있음을 보여줍니다.

### 체비셰프 부등식의 직관적 의미

체비셰프 부등식은 어떤 분포에서든, 확률변수가 기대값에서 벗어나는 정도가 표준편차의 몇 배 이상인 경우 그 확률이 매우 작음을 보여줍니다. 즉, 확률변수는 대부분의 경우 기대값 근처에 있으며, 기대값에서 멀리 떨어지는 경우의 확률은 분산에 반비례하여 감소한다는 의미입니다.



```python
import numpy as np
import matplotlib.pyplot as plt

# 체비셰프 부등식을 시각화하기 위한 값 설정
mu = 0  # 기대값
sigma = 1  # 표준편차
k_values = np.linspace(1, 5, 100)  # k값: 표준편차의 배수

# 체비셰프 부등식: P(|X - mu| >= k*sigma) <= 1 / k^2
chebyshev_upper_bound = 1 / k_values**2

# 시각화
plt.figure(figsize=(8, 6))
plt.plot(k_values, chebyshev_upper_bound, label=r'$\frac{1}{k^2}$', color='blue', lw=2)

# 그래프 레이블 설정
plt.title('Chebyshev Inequality Visualization', fontsize=16)
plt.xlabel(r'$k$', fontsize=14)
plt.ylabel(r'$P(|X - \mu| \geq k\sigma)$', fontsize=14)
plt.grid(True)

# 범례 추가
plt.legend(fontsize=12)

# 그래프 표시
plt.show()
```
